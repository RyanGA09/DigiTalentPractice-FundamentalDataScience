```python
# Load CSV file (update the path if needed)
data = pd.read_csv('./data/Data_Nasabah.csv', delimiter=';')

# Display the first five rows
print("\nFirst Five Rows of the Data:")
print(data.head())
```

```python
# Show data types and encoding info
print("\nData Information:")
print(data.info())
```

```python
# Descriptive statistics for numerical data
print("\nDescriptive Statistics:")
print(data.describe())
```

```python
# Display the number of unique values in each column
print("\nNumber of Unique Values per Column:")
for column in data.columns:
    print(f"{column}: {data[column].nunique()}")
```

```python
# Numerical Data Distribution Analysis
numerical_features = data.select_dtypes(include=['number']).columns
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[feature], kde=True)
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Correlation Matrix Analysis (only numerical features)
numerical_data = data.select_dtypes(include=['number'])  
correlation_matrix = numerical_data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Categorical Data Analysis
categorical_features = data.select_dtypes(include=['object']).columns
for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=feature, data=data)
    plt.title(f"Count of {feature}")
    plt.xticks(rotation=45)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()
```

```python
# Missing Values Analysis
missing_values = data.isnull().sum()
print("\nMissing Values per Column:")
print(missing_values)
```

```python
# Outlier Detection using boxplots
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=data[feature])
    plt.title(f"Outlier Detection for {feature}")
    plt.xlabel(feature)
    plt.show()
```

```python
# Specific column distribution analysis (e.g., 'umur')
plt.figure(figsize=(8, 6))
sns.histplot(data['umur'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Umur')
plt.ylabel('Frequency')
plt.show()
```

```python
# Relationship analysis between 'umur' and 'pendapatan'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='umur', y='pendapatan', data=data)
plt.title('Relationship between Age and Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
# Distribution of 'jenis_kelamin'
plt.figure(figsize=(6, 4))
sns.countplot(x='jenis_kelamin', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Relationship between 'jenis_kelamin' and 'pendapatan'
plt.figure(figsize=(8, 6))
sns.boxplot(x='jenis_kelamin', y='pendapatan', data=data)
plt.title('Relationship between Gender and Income')
plt.xlabel('Gender')
plt.ylabel('Income')
plt.show()
```

## Data Quality Verification

```python
# Deep and Detailed Data Validation

# 1. Check Data Consistency
# --------------------------

# Example: Check that 'umur' (age) column only contains positive values
print("\nValidating 'umur' (age) column:")
if data['umur'].min() < 0:
    print("⚠️ Found negative values in 'umur'. Further inspection is needed.")
else:
    print("✅ 'umur' contains only positive values.")

# Check that 'jenis_kelamin' contains only expected categories
print("\nValidating 'jenis_kelamin' (gender) column:")
invalid_gender = data['jenis_kelamin'].apply(lambda x: x not in ['Laki-Laki', 'Perempuan']).sum()
if invalid_gender > 0:
    print(f"⚠️ Found {invalid_gender} invalid entries in 'jenis_kelamin'.")
else:
    print("✅ All entries in 'jenis_kelamin' are valid.")
```
